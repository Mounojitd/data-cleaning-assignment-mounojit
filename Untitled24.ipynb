```python
import zipfile
import os

# Define paths
zip_path = "Cleaning of Data & Merging into single excel.zip"
extract_path = "/mnt/data/extracted_files"

# Extract zip file
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

# List extracted files and folders
extracted_files = []
for root, dirs, files in os.walk(extract_path):
    for file in files:
        extracted_files.append(os.path.join(root, file))

extracted_files[:10]  # Show a sample of the first 10 extracted files for verification
```

```python
import os
import pandas as pd

# Define folder path
folder_path = "Downloads/Cleaning of Data & Merging into single excel/Payout Summary & Order Level Sales"

# Create an empty list to store DataFr
merged_data = []
for filename in os.listdir(folder_path):
    if filename.endswith(".xlsx"):
        file_path = os.path.join(folder_path, filename)
        try:
            df = pd.read_excel(file_path)
            df["Source_File"] = filename 
            merged_data.append(df)
        except Exception as e:
            print(f"Error reading {filename}: {e}")
final_sales_df = pd.concat(merged_data, ignore_index=True)
print("Merged Sales Data:")
print(final_sales_df.head())
final_sales_df.to_excel("merged_sales_data.xlsx", index=False)
```

```python
import os
import pandas as pd
folder_path = "Downloads/Cleaning of Data & Merging into single excel/Payout Summary & Order Level Sales"
merged_data = []

for filename in os.listdir(folder_path):
    if filename.endswith(".xlsx"):
        file_path = os.path.join(folder_path, filename)
        try:
            df = pd.read_excel(file_path, sheet_name="Order Level", skiprows=1)
            df["Source_File"] = filename
            merged_data.append(df)
        except Exception as e:
            print(f"Error reading {filename}: {e}")
final_sales_df = pd.concat(merged_data, ignore_index=True)
final_sales_df.to_excel("merged_sales_data_cleaned.xlsx", index=False)
print("Final Cleaned Sales Data:")
print(final_sales_df.head())
```

```python
import os
import pdfplumber
import pandas as pd

pdf_folder = "Downloads/Cleaning of Data & Merging into single excel/Commission Invoices"
all_dataframes = []

for filename in os.listdir(pdf_folder):
    if filename.endswith(".pdf"):
        file_path = os.path.join(pdf_folder, filename)
        try:
            with pdfplumber.open(file_path) as pdf:
                for page in pdf.pages:
                    table = page.extract_table()
                    if table:
                        raw_headers = table[0]
                        headers = []
                        for i, col in enumerate(raw_headers):
                            if not col or col.strip() == "":
                                headers.append(f"Column_{i+1}")
                            elif col in headers:
                                headers.append(f"{col}_{i+1}")
                            else:
                                headers.append(col.strip())

                        df = pd.DataFrame(table[1:], columns=headers)
                        df["Source_File"] = filename
                        all_dataframes.append(df)
        except Exception as e:
            print(f"Error reading {filename}: {e}")

if all_dataframes:
    final_df = pd.concat(all_dataframes, ignore_index=True)
    final_df.to_excel("commission_invoices_merged.xlsx", index=False)
    print("PDF conversion complete. Saved to 'commission_invoices_merged.xlsx'")
    print(final_df.head())
else:
    print("No tables found in PDFs.")
```
